## Aggregation methods to do:
1. Overall volume of positive sentiment by candidate, for the entire collection period
2. Overall volume of positive sentiment by candidate, for the entire collection period, aggregated to the per-user level (so just one sentiment score per user)
[do weekly analyses seperately ] 3. Overall volume of positive sentiment by candidate, segmented by week, user-level?
4. Overall volume of positive sentiment by candidate, at the user-level, weighting by population density (i.e. up-weighting pos sent from under-represented areas and down-weighting pos sent from over-represented areas)
5. Same as 4., but segmented by week
6. Old method, i.e.: take the n=1000 hand-labelled sample for each state. Build a synthetic distribution of these scores where the n more or less matches the number of folks that'll vote in this election. From the probability density distribution for each distribution, generate at random a sentiment score for either candidate. Assign a 'vote' based on which score is higher. 
7. Old method, but for the entire propagated sample instead of just the small amount of labelled data
8. Old method, but with adjustment by location, for the full propagated sample. 
9. Net sentiment - not quite sure how this can be done, because if the net sentiment for a given candidate is negative, how would that transpire to any individual votes? 
10. Repeat both method 1-3 and method 5-7, but exclude folks that were labelled by pabloscores as being 'conservative', i.e. only moderates and liberals.

```r
nh_user_level_df <- NULL
```

```r
for(i in 1:length(unique(nh_df$user_id))){
    # subset to individual users
    tmp_df <- nh_df %>% 
        filter(user_id == unique(nh_df$user_id)[i])
    # generate mean score
    mean_hillary <- ifelse(is.nan(mean(tmp_df$hillary_preds_log, na.rm = T)), NA, 
                                  mean(tmp_df$hillary_preds_log, na.rm = T))               
    mean_bernie <- ifelse(is.nan(mean(tmp_df$bernie_preds_log, na.rm = T)), NA, 
                                  mean(tmp_df$bernie_preds_log, na.rm = T))
    tmp_df <- tmp_df %>% 
        select(all_of(required_cols))
    tmp_df$mean_hillary <- mean_hillary
    tmp_df$mean_bernie <- mean_bernie
    # merge on to nh_user_level_df
    newline_df <- tmp_df[nrow(tmp_df),]
    nh_user_level_df <- rbind(nh_user_level_df, newline_df)
}
```

```r
nh_b_pos_sent_user <- length(nh_user_level_df$mean_bernie[nh_user_level_df$mean_bernie>5])
nh_h_pos_sent_user <- length(nh_user_level_df$mean_hillary[nh_user_level_df$mean_hillary>5])
nh_b_pos_sent_perc_user <- nh_b_pos_sent_user/(nh_b_pos_sent_user+nh_h_pos_sent_user)*100
nh_h_pos_sent_perc_user <- nh_h_pos_sent_user/(nh_b_pos_sent_user+nh_h_pos_sent_user)*100
print(paste0('Percentage Bernie in NH: ', nh_b_pos_sent_perc_user))
print(paste0('Percentage Hillary in NH: ', nh_h_pos_sent_perc_user))
mae_02_nh <- mean(abs(c((nh_b_pos_sent_perc_user-real_bernie_nh), (nh_h_pos_sent_perc_user-real_hillary_nh))))
print(paste0('Mean absolute error for this model in the NH primary: ', mae_02_nh))
```

## 3. Same as in 2., but weighted by tweeter's population density  
For this, I will utilise the 'pop_quantiles' variable in the _user_level_df_ data frame, so as to more accurately reflect the geographic makeup of a given state.  
There are a few ways of doing this. I think I'll focus on two possible options: 

- duplicating (randomly) cases that are under-represented while removing (randomly) cases that are over-represented, so as to be left with a (geographically) representative sample of users
- somehow simulating users and their scores but making sure that the simulation information is drawn from the actual figures which we have in the data. However, I think it'll be best to do this later on, as it is closer to the original research design we came up with for this paper. 

Regardless, I'll start by reading in the population distribution dataframe. 

```r
state_level_representativeness_df <- read.csv("~/Desktop/SENTIMENT/data/users_with_place_proportions/percentages_per_quantile.csv",
                                             stringsAsFactors = FALSE)
```

```r
state_level_representativeness_df
```

Now, let's calculate the weighting factor for each of the quantiles, and then add it on to the DF.

```r
rows <- list(c(1,2), c(3,4), c(5,6))
quantiles <- c('q1', 'q2', 'q3', 'q4')
```

```r
weighting_factors_df <- NULL
for(i in 1:length(quantiles)){
    newcol <- NULL
    for(j in 1:length(rows)){
        the_factor <- (state_level_representativeness_df[quantiles[i]][rows[[j]][1],]/
                        state_level_representativeness_df[quantiles[i]][rows[[j]][2],])

        newcol <- c(newcol, NA, the_factor)
    }
    weighting_factors_df <- as.data.frame(cbind(weighting_factors_df, newcol))
}
colnames(weighting_factors_df) <- c('weight_q1', 'weight_q2', 'weight_q3', 'weight_q4')
```

```r
state_level_representativeness_df <- cbind(state_level_representativeness_df, weighting_factors_df)
```

```r
state_level_representativeness_df
```

Now, let's generate a weighted dataset for each of the state-level dataframes. This will mean sampling rows which represent users that live in a location which is in population quantile i, **with replacement**. The new dataframe will then have n rows for each population quantile, whereby n represents the number of rows in the current (unweighted) user-level dataframe for a given quantile, multiplied by the weighting factor (in state_level_representativeness_df).

1. Make upsampled DF

```r
nrow(nh_user_level_df)
print('ns per pop-quantile in sample')
table(nh_user_level_df$pop_quantiles)
print('target ns per pop-quantile in sample')
target_rownums_nh_df <- ceiling(table(nh_user_level_df$pop_quantiles)*state_level_representativeness_df[4,8:11]) 
# weighted up to nearest integer
target_rownums_nh_df
```

```r
nh_user_level_weighted_df <- NULL
for(i in 1:length(target_rownums_nh_df)){
    # go through each of the 
    tmp_df <- nh_user_level_df %>% 
        filter(pop_quantiles==as.character(i))
    # generate sampling indeces
    new_indeces <- sample(x = 1:nrow(tmp_df), size = target_rownums_nh_df[1,i], replace = TRUE)
    # new lines for data frame
    newlines_df <- tmp_df[new_indeces,]
    nh_user_level_weighted_df <- rbind(nh_user_level_weighted_df, newlines_df)
}
```

```r
prop.table(table(nh_user_level_weighted_df$pop_quantiles))
head(nh_user_level_weighted_df)
```

## 4. Old method -- building a synthetic dataset of simulated 'candidate approval scores'
See the old draft and code for this paper. What we do here is the following: 
- start with reading in **only** the hand-labelled data for each of the states. 
- have a look at the kernel density plots for these distributions
- from these distributions, sample n randomly (and continuously) drawn scores from **each** candidate's distribution, whereby n represents the expected number of voters in the respective election. 
- these numbers are now aggregated: for every 'voter', the vote goes to the candidate with the relatively higher synthetic 'approval score'

### NOTE: here, scores go from -5 to +5

What we're trying to do now is to to repeat the same ultra-basic method as shown in 7., but we're going to stratify the small samples of n=1000 of labelled tweets for each of the states so that the population density distribution reflects the real-world distribution.  
1. establish pop distribution within the sample
2. generate a new, pop-distribution weighted sample
3. repeat the same method as above. 

### A. New Hampshire

```r
table(nh_handlabelled_df$pop_quantiles)
```

```r
# population distribution in the labelled sample
nh_sample_distrib <- (prop.table(table(nh_handlabelled_df$pop_quantiles))*100)
nh_real_distrib <- state_level_representativeness_df[3,][4:7]
# calculate factors
nh_weighting_factors <- nh_real_distrib/nh_sample_distrib
target_rownums_nh_justlabelled_df <- ceiling(nh_weighting_factors*table(nh_handlabelled_df$pop_quantiles))
# target number of rows
target_rownums_nh_justlabelled_df
```

```r
nh_handlabelled_weighted_df <- NULL
for(i in 1:length(target_rownums_nh_justlabelled_df)){
    # go through each of the 
    tmp_df <- nh_handlabelled_df %>% 
        filter(pop_quantiles==i)
    # generate sampling indeces
    new_indeces <- sample(x = 1:nrow(tmp_df), size = target_rownums_nh_justlabelled_df[1,i], replace = TRUE)
    # new lines for data frame
    newlines_df <- tmp_df[new_indeces,]
    nh_handlabelled_weighted_df <- rbind(nh_handlabelled_weighted_df, newlines_df)
}
```

```r
nh_b_pos_sent_justlabelled_weighted <- length(nh_handlabelled_weighted_df$bernie_score[nh_handlabelled_weighted_df$bernie_score>0])
nh_h_pos_sent_justlabelled_weighted <- length(nh_handlabelled_weighted_df$hillary_score[nh_handlabelled_weighted_df$hillary_score>0])
nh_b_pos_sent_justlabelled_weighted_perc <- (nh_b_pos_sent_justlabelled_weighted/(nh_b_pos_sent_justlabelled_weighted+nh_h_pos_sent_justlabelled_weighted))*100
nh_h_pos_sent_justlabelled_weighted_perc <- (nh_h_pos_sent_justlabelled_weighted/(nh_b_pos_sent_justlabelled_weighted+nh_h_pos_sent_justlabelled_weighted))*100
print(paste0('Percentage (just hand-labelled, weighted) Bernie in NH: ', nh_b_pos_sent_justlabelled_weighted_perc))
print(paste0('Percentage (just hand-labelled, weighted) Hillary in NH: ', nh_h_pos_sent_justlabelled_weighted_perc))
mae_08_nh <- mean(abs(c((nh_b_pos_sent_justlabelled_weighted_perc-real_bernie_nh), (nh_h_pos_sent_justlabelled_weighted_perc-real_hillary_nh))))
print(paste0('Mean absolute error for this model in the NH primary: ', mae_08_nh))
```

## WHAT EVEN WERE THE KEY QUESTIONS:
- Is there any advantage to working with granular sentiment versus categorical sentiment? Intuitively you'd think, yes, because there's so much more information there. 
- The state of the art in this field precludes that sentiment scores should be propagated to a much larger sample of tweets that can feasibly be humanly labelled by researchers using machine learning. However, in an applied social science setting like this one, is there any tangible benefit from doing so? 
- Previous research has stated that weighting, de-biasing and post-stratification of tweet samples could help improve accuracy and reproducibility of twitter-based election forecasting. Is this the case, empirically, in these examples? 

# MAIN TAKEAWAYS: 
- you can fudge around with the data as much as you want, but at the end of the day, the established method (pos sent volume) works best
- what are we actually observing here? what are people talking about? is it national politics or is it local politics? are the folks that tweet actually going to vote? who are the folks that tweet? 
- interestingly, location -- typically a strong predictor of how folks are going to vote -- gives us very, very little change here, in fact in the case of MA and SC it worsens the model performance. 
    - so, folks that were on twitter in 2016 and lived in some of these states, rurally, are very much not like your average rural folks
- clearly, this method is not really fit for purpose. 
- but: there are interesting things that we can learn here. hillary was not very popular, at a point when she should really have been. this is abundantly clear when looking at these data - from which ever angle one wishes to do so
- did bernie manage to perfectly mobilise twitter folks in new hampshire? why is it so accurate there? probably just coincidence? 
- i argue that this might underline an important shortcoming with the state of the art of twitter-based social science. sampling. simply sampling data based on the existence of X isn't a useful approach. A better approach is likely to randomly sample from a given target population, and then reduce this down to X. 

- ultimately it would be interesting to model which factors would be needed to complement the twitter data as is, and to what degree they should be included in a model in order for these data to be used in a predictive manner -- BUT this might be a little too much for this paper, late as it already is

**RE generating ordinal vs categorical sentiment scores**
- there doesn't appear to be too much of an advantage of labelling tweets like this vs just positive-neutral-negative


**RE machine propagating ordinal sentiment scores to tweets**
- it's not that reliable, it seems, and it appears as if just sticking with hand-labelled data does a better job of it


## CAVEATS:
- we're using fairly small sample sizes here, and we're working with primaries rather than, e.g., the presidential election. so, perhaps this paper also clearly illustrates the limits of public opinion destillation from tweets in the electoral context, rather than necessarily ruling this sentiment-assisted approach out as a whole.
